```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("all_result_for_tresh1-255_pivoted.csv")
```

```python
df['correct_bacteria_count'] = df["Krzywe"]+df["Pałeczki"]+	df["Ziarniaki"]
```

```python
df.head()
```

```python
# Set threshold as x-axis
x = df["threshold"]

# Plot each bacteria type
bacteria_types = df.columns.tolist()
bacteria_types.remove("threshold")  # exclude threshold column

plt.figure(figsize=(10, 6))
for btype in bacteria_types:
    plt.plot(x, df[btype], marker='o', label=btype)

plt.xlabel("Threshold")
plt.ylabel("Count")
plt.title("Bacteria counts vs Threshold")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
